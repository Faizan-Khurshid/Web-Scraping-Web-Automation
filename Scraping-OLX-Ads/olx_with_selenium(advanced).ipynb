## Scraping the Site

```python
driver = webdriver.Chrome("E:\softwares\chromedriver_win32\chromedriver.exe")
driver.get("https://www.olx.com.pk/item/suziki-swift-car-urgent-sall-iid-1033466359") #swift cars on olx
# while True:
#     try:
#         driver.implicitly_wait(5)
#         driver.find_element_by_xpath('//button[@data-aut-id="btnLoadMore"]').click() # load all ads
#     except:
#         break
cars_title = driver.find_elements_by_xpath('//span[@data-aut-id="itemTitle"]')
cars_detail = driver.find_elements_by_xpath('//span[@data-aut-id="itemDetails"]')
cars_loc = driver.find_elements_by_xpath('//span[@data-aut-id="item-location"]')
# # ad_post_date = driver.find_element_by_xpath('//span[@data-aut-id="item-location"]')
cars_price = driver.find_elements_by_xpath('//span[@data-aut-id="itemPrice"]')

# driver.close()
```

```python
cars_title
```

## Display Data

```python
data = []
for title, detail, loc, price in zip(cars_title, cars_detail, cars_loc, cars_price):
    data.append([title.text, detail.text, loc.text, price.text])
```

```python
#display data
print(tabulate(data, headers=["Title", "Details", "Location", "Price"]))
```

```python
# opening the csv file in 'a+' mode
file = open('olx_swift.csv', 'a+', newline ='', encoding="utf-8")
  
# writing the data into the file
with file:    
    write = csv.writer(file)
    write.writerows(data)
```
